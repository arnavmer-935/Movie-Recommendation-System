```python
movies = pd.read_csv("/content/drive/MyDrive/AI ML/CSV Data Files/tmdb_5000_movies.csv")
credits = pd.read_csv("/content/drive/MyDrive/AI ML/CSV Data Files/tmdb_5000_credits.csv", encoding = "latin1")
```

```python
movies = movies.merge(credits, on = "title")
movies.head(2)
```

```python
movies.shape
```

```python
movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
movies.head(2)
```

```python
movies.iloc[0].genres #Obtains entry in first row of genres column
```

```python
import json #Used to access individual dictionaries enclosed in a string

def convert(text):
  dictlist = json.loads(text)
  L = []
  for i in dictlist:
    L.append(i["name"])
  return L
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
movies.head(2)
```

```python
movies.iloc[0].cast
```

```python
def castconvert(text):
  castlist = json.loads(text)
  L = []
  castcount = 0
  for i in castlist:
    if castcount < 3:
      L.append(i["name"])
    castcount += 1
  return L
```

```python
movies["cast"] = movies["cast"].apply(castconvert)
movies.head(2)
```

```python
movies.iloc[0].cast
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
def director(text):
  dl = json.loads(text)
  L = []
  for i in dl:
    if i["job"] == "Director":
      L.append(i["name"])

  return L
```

```python
movies["crew"] = movies["crew"].apply(director)
movies.head(2)
```

```python
movies.head(2)
```

```python
movies["genres"] = movies["genres"].apply(convert)
movies.head(2)
```

```python
def spr(L):
  L1 = []
  for word in L:
    L1.append(word.replace(" ", ""))

  return L1
```

```python
movies["genres"] = movies["genres"].apply(spr)
movies["keywords"] = movies["keywords"].apply(spr)
movies["cast"] = movies["cast"].apply(spr)
movies["crew"] = movies["crew"].apply(spr)
```

```python
movies.head(2)
```

```python
movies.dropna(inplace = True)
```

```python
def splitter(text):
  wordL = text.split()
  return wordL
```

```python
movies.head(2)
```

```python
movies["overview"] = movies["overview"].apply(splitter)
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
movies.head()
```

```python
movies.iloc[0].tags
```

```python
final = movies[["movie_id", "title", "tags"]]
final.head()
```

```python
def joiner(text):
  st = " ".join(text)
  return st
```

```python
final["tags"] = final["tags"].apply(joiner)
final.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec = CountVectorizer(max_features = 5000)
```

```python
tag_vector = vec.fit_transform(final["tags"]).toarray()
tag_vector.shape
```

```python
print(len(tag_vector[0]))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(tag_vector)
similarity
```

```python
vec = similarity[0]
vec
```

```python
reversed_similarity = np.sort(vec)[::-1]
reversed_similarity
```

```python
def recommend(movie):
  index = final[final["title"] == movie].index[0]
  distances = sorted(list(enumerate(similarity[index])), reverse = True, key = lambda x: x[1])
  for i in distances[1:6]:
    print(final.iloc[i[0]].title)
```

```python
recommend("Harry Potter and the Half-Blood Prince")
```
